##**Probabilidade Simples**

- Qual a probabilidade de selecionar um segurado e essa pessoa ter a reclamação de seguro aprovada?  

````python
ProbAprovado = round(len(CarInsurance[CarInsurance['OUTCOME'] == 1.0]) / len(CarInsurance), 4)
print ('A probabilidade de uma solicitação ser aprovada é de:' + str(round(ProbAprovado*100,4)) + '%')
````


- Qual a probabilidade de selecionar uma pessoa e essa pessoa ser mulher?

````python
ProbFe = round(len(CarInsurance[CarInsurance['GENDER'] == 'female']) / len(CarInsurance),4)
ProbFe
print ('A probabilidade de uma mulher ter contrato com o seguro de veículos é:' + str(ProbFe*100) + '%')
````

##**Probabilidade União**

- Qual a probabilidade do segurado ter se envolvido em acidente ou ter histórico de direção sob efeito de álcool?

````python
AciMul = len(CarInsurance[(CarInsurance['PAST_ACCIDENTS']>0 | (CarInsurance['DUIS']>0))])
ProbAciMul = round(AciMul/len(CarInsurance), 4)
print('A probabilidade de ter se envolvido em acidente ou ter tido multa por velocidade é: ' + str(ProbAciMul*100) + '%')
````

##**Probabilidade Intersecção**

- Qual a probabilidade de ser mulher e ter se envolvido em acidentes anteriores?

````python
AciFe = len(CarInsurance[(CarInsurance['PAST_ACCIDENTS']>0) & (CarInsurance['GENDER'] == 'female')])
ProbAciFe = AciFe/len(CarInsurance)
print('A probabilidade de ser mulher e ter se envolvido em acidentes anteriores: ' + str(ProbAciFe*100) + '%')
````

- Qual a probabilidade de ser homem e ter se envolvido em acidentes anteriores?

````python
AciMa = len(CarInsurance[(CarInsurance['PAST_ACCIDENTS']>0) & (CarInsurance['GENDER'] == 'male')])
ProbAciMa = AciMa/len(CarInsurance)
print('A probabilidade de ser homem e ter se envolvido em acidentes anteriores: ' + str(ProbAciMa*100) + '%')
````

##**Probabilidade do Complemento**

Qual a probabilidade de ser uma cliente mulher e não ter se envolvido em acidente?

````python
ProbFe = round(len(CarInsurance[CarInsurance['GENDER'] == 'female']) / len(CarInsurance),4)
ProbFeNaoAci = ProbFe - ProbAciFe
print('A probabilidade de ser uma cliente mulher e não ter se envolvido em acidente é de : '+ str(round(ProbFeNaoAci*100, 4)) + '%')
````

##**Probablidiade da Diferença**

Qual a probabilidade do cliente ter se envolvido em acidente mas não ter histórico de direção sob efeito de álcool?  

````python
AciNaoAlc = len(CarInsurance[CarInsurance['PAST_ACCIDENTS'] >0]) - len(CarInsurance[CarInsurance['DUIS']>0])
ProbAciNaoAlc = AciNaoAlc/len(CarInsurance)
print('A probabilidade de ter se envolvido em acidente mas não ter histórico de direção sob efeito de álcool é:' + str(round(ProbAciNaoAlc*100,4)) + '%')
````

#**Probabilidade Condicional**

Qual a probabilidade de ter se envolvido em acidente anterior sabendo que é uma mulher?

````python
ProbCond = ProbAciFe/ProbFe
print('A probabilidade de ter se envolvido em acidente anterior sabendo que é uma mulher é : '+ str(round(ProbCond*100, 4)) + '%')
````
